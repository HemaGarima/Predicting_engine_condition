The Dataset contains 19535 rows and 7 columns. The values are used to analyse the condition of the engine in a car. The above codes give a description about the columns present in dataset.

Column 1: Engine rpm-  RPM indicates how many times the engine's crankshaft completes a full rotation within a minute.

Column 2: Lub oil pressure- Lubricating oil pressure, often referred to as "lub oil pressure," is a measure of the force with which oil is supplied to lubricate and protect the moving parts within an engine or machinery.

Column 3: Fuel pressure- Fuel pressure is a measure of the force with which fuel is delivered to the engine's fuel system.

Column 4: Coolant pressure- Coolant pressure refers to the pressure exerted by the liquid coolant circulating within the cooling system of an engine or a machine.

Column 5: lub oil temp- Lubricating oil temperature, often abbreviated as "lub oil temp," refers to the temperature of the oil used to lubricate the moving parts within an engine or machinery.

Column 6: Coolant temp-  Coolant temperature, often abbreviated as "coolant temp," refers to the temperature of the liquid coolant circulating through the cooling system of an engine.

Column 7: Engine Condition- "Engine condition" is a general term used to describe the overall state or health of an engine. It encompasses various factors, including the engine's performance, efficiency, wear and tear, and any potential issues or problems.

Are there any outliers in the lub oil temp , coolant temp and lub oil pressure? If so, how many? Can you identify the values of the outliers?

As observable, there are many outliers in the data. We will do an outlier analysis using Z score for the coolant temp column and IQR method for the rest. Lets check the no. of outliers in each column

```python
for i in ['Engine rpm', 'Lub oil pressure', 'Fuel pressure','Coolant pressure','lub oil temp']:
    percentile25 = engine[i].quantile(0.25)
    percentile75 = engine[i].quantile(0.75)

    iqr = percentile75 - percentile25

    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr
    oulier_count=len(engine[engine[i]>upper_limit])+len(engine[engine[i]<lower_limit])
    print("The no. of outliers in", i, 'is', oulier_count)
    engine[i] = np.where(
    engine[i] > upper_limit, upper_limit, np.where(engine[i] < lower_limit, lower_limit,engine[i]))
```

```python
upper_limit_cooltemp = engine['Coolant temp'].mean() + 3*engine['Coolant temp'].std()
lower_limit_cooltemp = engine['Coolant temp'].mean() - 3*engine['Coolant temp'].std()
engine['Coolant temp'] = np.where(
    engine['Coolant temp']>upper_limit_cooltemp,
    upper_limit_cooltemp,
    np.where(
        engine['Coolant temp']<lower_limit_cooltemp,
        lower_limit_cooltemp,
        engine['Coolant temp']
    )
)
```

```python
sns.boxplot(engine['Coolant temp'])
```

```python
sns.boxplot(engine['Lub oil pressure'])
```

 **SUMMARY**
 The above code shows the number of outliers in lub oil temp , coolant temp and lub oil pressure. It is represented using Boxplot.

Using Z score and IQR method we have reduced the number of outliers which are represented using Boxplot (4,5,6)).

<-----------------------Question 3----------------------------->

Show the significance of each column given in the Dataset. Show which column values have a significant effect on Engine Condiiton column.


```python
import scipy.stats as stats
```

```python
print('The F-statistic')
print(' It is a measure of the ratio of between-group variability to within-group variability.' )
print(' A larger F-statistic indicates a stronger evidence against the null hypothesis of equal means across groups.')
```

```python
print('P_Value')
print('P_Values indicates the level of statistical significance of the observed group differences.')
print('If the p-value is below a predetermined significance level , it suggests that the observed group differences are unlikely to occur by chance alone, and the null hypothesis can be rejected.')
```

```python
groups=[]
for group_name, group_data in df.groupby('Engine rpm')['Engine Condition']:
  groups.append(group_data)
f_stat, p_value = stats.f_oneway(*groups)

# Print the ANOVA results
print("ANOVA results:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value > 0.05:
  print('There is a significant effect on the Engine rpm based on Engine Condition.')
else:
  print('There is no significant effect on the Engine rpm based on Engine Condition.')
```

```python
groups=[]
for group_name, group_data in df.groupby('Lub oil pressure')['Engine Condition']:
  groups.append(group_data)
f_stat, p_value = stats.f_oneway(*groups)

# Print the ANOVA results
print("ANOVA results:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value > 0.05:
  print('There is a significant effect on the Lub oil pressure based on Engine Condition.')
else:
  print('There is no significant effect on the Lub oil pressure based on Engine Condition.')
```

```python
groups=[]
for group_name, group_data in df.groupby('Coolant pressure')['Engine Condition']:
  groups.append(group_data)
f_stat, p_value = stats.f_oneway(*groups)

# Print the ANOVA results
print("ANOVA results:")
print("p-value:", p_value)
if p_value > 0.05:
  print('There is a significant effect on the Coolant pressure based on Engine Condition.')
else:
  print('There is no significant effect on the Coolant pressure based on Engine Condition.')
```

```python
groups=[]
for group_name, group_data in df.groupby('Fuel pressure')['Engine Condition']:
  groups.append(group_data)
f_stat, p_value = stats.f_oneway(*groups)

# Print the ANOVA results
print("ANOVA results:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value > 0.05:
  print('There is a significant effect on the Fuel pressure based on Engine Condition.')
else:
  print('There is no significant effect on the Fuel pressure based on Engine Condition.')
```

```python
groups=[]
for group_name, group_data in df.groupby('lub oil temp')['Engine Condition']:
  groups.append(group_data)
f_stat, p_value = stats.f_oneway(*groups)

# Print the ANOVA results
print("ANOVA results:")
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value > 0.05:
  print('There is a significant effect on the lub oil temp based on Engine Condition.')
else:
  print('There is no significant effect on the lub oil temp based on Engine Condition.')
```
